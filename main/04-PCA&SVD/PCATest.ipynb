```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
iris = load_iris()
iris.data.shape
```

```python
X = iris.data
y = iris.target
pd.DataFrame(X)
```

```python
#从四维降到二维
pca = PCA(n_components=2)
pca = pca.fit(X)
X_dr = pca.transform(X)
X_dr
```

```python
plt.figure()
plt.scatter(X_dr[:,0],X_dr[:,1])
plt.show()
```

```python
X_dr[y == 0,0]
```

```python
ylabels = set(y.tolist())
colors = ['red', 'black', 'orange']
```

```python
plt.figure()
for i in ylabels:
    plt.scatter(X_dr[y==i,0],X_dr[y==i,1],c=colors[i],alpha = 0.7,label = iris.target_names[i])
plt.legend()
plt.title('PCA of IRIR dataset')
plt.show()
```

```python
pca.explained_variance_ #降维后的，新特征对应的方差
```

```python
pca.explained_variance_ratio_ #降维后的，新特征对应方差相比原来特征矩阵的方差贡献率
```

```python
## 累计可解释性方差贡献率曲线
pca_line=PCA().fit(X)
plt.plot([1,2,3,4],np.cumsum(pca_line.explained_variance_ratio_))
plt.xticks([1,2,3,4]) 
plt.xlabel("number of components after dimension reduction")
plt.ylabel("cumulative explained variance ratio")
plt.show()
```

```python
###这个地方显示的是三维的时候，模型方差贡献率累加已经接近0.993了
### 我们尝试一下建模
pca = PCA(3)
pca = pca.fit(X)
X_dr_3 = pca.transform(X)
pd.DataFrame(X_dr_3).head()
```

```python

ax = plt.axes(projection='3d')
for i in ylabels:
    ax.scatter3D(X_dr_3[y==i,0],X_dr_3[y==i,1],X_dr_3[y==i,2],c=colors[i])
```

```python
###最大似然估计方式自选超参数n_components,不要我们苦恼地挑选K了
pca_mle = PCA(n_components='mle')
pca_mle = pca_mle.fit(X)
X_mle = pca_mle.transform(X)
X_mle.shape #自动帮我们选了3个维度
```

```python
pca_mle.explained_variance_ratio_.sum() #超参数累加方差贡献率达0.994
```

```python
# 按信息量占比自选超参数
# 输入[0,1]之间的浮点数，并且让参数svd_solver =='full'，表示希望降维后的总解释性方差占比大于n_components
# 指定的百分比，即是说，希望保留百分之多少的信息量

pca_f = PCA(n_components=0.99,svd_solver='full')
pca_f = pca_f.fit(X)
X_dr_f =pca_f.transform(X)
pd.DataFrame(X_dr_f).head() #0.97返回二维，0.99返回3维
```

```python
pca_f.explained_variance_ratio_.sum()
```

```python
pca_f.components_ #Vk【3*4】即【k*n】k是新特征数，n是原特征矩阵的特征数
```

```python
###人脸识别数据中，对VK的运用
import numpy as np
import os
from skimage import io
from skimage import transform


def read_imagelist(filelist):
    # brief：从列表文件中，读取图像数据到矩阵文件中
    # param： filelist 图像列表文件
    # return ：4D 的矩阵

    fid = open(filelist)
    lines = fid.readlines()
    test_num = len(lines)
    fid.close()
    X = []
    
    for line in lines:
        word = line.split('\n')
        filename = 'E:/openData/faceData/lfw/lfw/lfw' + word[0]
        if not os.path.exists(filename):
            continue
        im1 = io.imread(filename,as_gray=True)
        X.append(im1.tolist())
        
    return X


def read_labels(labelfile):
    fin = open(labelfile)
    lines = fin.readlines()
    labels = np.empty((len(lines),))
    k = 0;
    for line in lines:
        labels[k] = int(line)
        k = k + 1;
    fin.close()
    return labels


left = read_imagelist('E:/openData/faceData/lfw/LFW数据读取（两种方法）/little_left.list')
```

```python
###用PCA做噪声过滤
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

digits = load_digits()
```

```python
digits.data.shape
```

```python
set(digits.target)
```

```python
#定义画图函数
def draw_digits(data):
    fig,axes = plt.subplots(4,10,figsize=(10,4),subplot_kw={'xticks':[],'yticks':[]})
    for i,ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8,8),cmap='binary')
    fig
draw_digits(digits.data)
```

```python
### 为数据加上噪声
np.random.RandomState(42)
noisy = np.random.normal(digits.data,2) #方差为2
draw_digits(noisy)
```

```python
## 降维并降噪
pca = PCA(0.5)
X_dr = pca.fit_transform(noisy)
X_dr.shape
```

```python
without_noise =pca.inverse_transform(X_dr)
draw_digits(without_noise)
```

```python
### 案例：PCA对手写数字数据集的降维
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv(r'C:\Users\Mypc\Desktop\菜菜\源文件\03-数据预处理与特征工程\digit recognizor.csv')
X= data.iloc[:,1:]
y = data.iloc[:,0]
X.shape
```

```python
### 画累计方差贡献率曲线
pca_line=PCA().fit(X)
pca_line.explained_variance_ratio_
plt.figure()
plt.plot(np.cumsum(pca_line.explained_variance_ratio_))
plt.xlabel("number of components after dimension reduction")
plt.ylabel("cumulative explained variance ratio")
plt.show()
```

```python
### 从图上可以看出100~200维比较不错
score=[]
for i in range(100,200,1):
    X_dr = PCA(i).fit_transform(X)
    once =cross_val_score(RFC(n_estimators=10,random_state = 0),X_dr,y,cv=5).mean()
    score.append(once)
plt.figure(figsize=[20,5])
plt.plot(range(100,200,1),score)
plt.show()

#======【TIME WARNING：2mins 30s】======#
```
